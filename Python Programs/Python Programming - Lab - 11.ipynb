### 01) WAP to demonstrate the use of Pie chart.

```python
ex = [0.2,0.0,0.0,0.0]
x = [10,20,30,40]
c = ["r","g","y","m"]
y = ["food","travel","accommodation","shopping"]
plt.pie(x,labels=y,colors = c,radius=1.2,shadow=True,autopct = "%0.1f%%",explode=ex,labeldistance=1.3,textprops={"fontsize":15,"color":"g"})
plt.title("Expenses",color="b",fontsize=15)
plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
x = ["python","c","c++","java"]
y = [85,70,60,82]
c = ["r","g","y","m"]
plt.bar(x,y,color = c,align="edge",edgecolor="r",alpha=0.5)
plt.title("Popularity Of Language",fontsize=15,color="g")
plt.xlabel("languages")
plt.ylabel("popularity in percentage")
plt.show()
```

```python
x = ["python","c","c++","java"]
y = [85,70,60,82]
c = ["r","g","y","m"]
plt.barh(x,y,color = c,align="edge",edgecolor="r",alpha=0.5)
plt.title("Popularity Of Language",fontsize=15,color="g")
plt.xlabel("languages")
plt.ylabel("popularity in percentage")
plt.show()
```

### 04) WAP to demonstrate the use of Histogram.

```python
import numpy as np
cpi = np.random.randint(0,10,100)
plt.hist(cpi,label="cpilist")
plt.legend()
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x = ["python","c","c++","java"]
y = [85,70,60,82]
c = ["r","g","y","m"]
plt.bar(x,y,color = c,align="edge",edgecolor="r",alpha=1)
plt.title("Popularity Of Language",fontsize=15,color="g")
plt.xlabel("languages")
plt.ylabel("popularity in percentage")

def labels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
labels(x,y)
plt.show()
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x=[1,2,3,4,5]
c = ["r","g","b","y","m"]
size = [20,40,15,70,50]
y= [2,4,6,8,10]
plt.scatter(x,y,color=c,s=size,marker="*")
plt.show()
```
